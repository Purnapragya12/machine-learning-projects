```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Purnapragya Sinha\Downloads\5c05f43d2351249927b8a3f3cc3e5ecf-8bd6144a87988213693754baaa13fb204933282d\5c05f43d2351249927b8a3f3cc3e5ecf-8bd6144a87988213693754baaa13fb204933282d\Mall_Customers.csv")
X = df[['Annual Income (k$)','Spending Score (1-100)']]

kmeans = KMeans(n_clusters=5)
df['Cluster'] = kmeans.fit_predict(X)

plt.scatter(X.iloc[:,0], X.iloc[:,1], c=df['Cluster'])
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```
